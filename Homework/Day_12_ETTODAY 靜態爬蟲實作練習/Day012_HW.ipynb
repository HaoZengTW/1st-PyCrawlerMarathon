### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
d = requests.get(url)

soup = BeautifulSoup(d.text, "html.parser")
for i in soup.find(class_= 'part_list_2').find_all('h3'):
    print(i.find(class_='date').text, i.find(class_='tag').text, '\n',i.find(target="_blank").text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
Etoday={}
for i in soup.find(class_= 'part_list_2').find_all('h3'):
    if i.find(class_='tag').text in Etoday:
        Etoday[i.find(class_='tag').text].append({'Time' : i.find(class_='date').text ,'Text': i.find(target="_blank").text})
    else:
        Etoday[i.find(class_='tag').text] = [{'Time' : i.find(class_='date').text ,'Text': i.find(target="_blank").text}]
```

```python
for i in Etoday:
    print('\n', i,'\n')
    for t in Etoday[i]:
        print(t)
```

### ③ 哪一個類別的文章最多

```python
title=''
count=0
for i in Etoday:
    if len(Etoday[i])>count:
        title = i
    #print(i, len(Etoday[i]))
print(title, )
```
